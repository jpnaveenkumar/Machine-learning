```python
#pd.to_numeric(data["imdbVotes"],errors='ignore')
for i in range(0,len(data['imdbVotes'])):
    #print(i,data.loc[i]["imdbVotes"])
    if(data.loc[i]["imdbVotes"]=='NaN' or data.loc[i]["imdbVotes"]=='-' or str(data.loc[i]["imdbVotes"])=='nan' or str(data.loc[i]["imdbVotes"])==''):
        data.loc[i]["imdbVotes"]=0
    else:
        value=str(data.loc[i]["imdbVotes"])
        if ',' in value:
            value=value.replace(',','')
        #print(value)
        #if(value!='nan'):
        data.loc[i]["imdbVotes"]=int(value)
```

```python
for i in range(0,len(data["Year"])):
    if '\x96' not in data.loc[i]["Year"]:
        #print(data.loc[i]["Year"])
        data.loc[i]["Year"]=int(str(data.loc[i]["Year"]))
    else:
        val=data.loc[i]["Year"]
        #print(val)
        if val.find('\x96')==(len(val)-1):
            data.loc[i]["Year"]=int(val[0:len(val)-1])
        else:
            ind=val.find('\x96')
            year1=int(val[0:ind])
            year2=int(val[ind+1:])
            data.loc[i]["Year"]=int((year1+year2)//2)
```
